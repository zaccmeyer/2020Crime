```python
import pyvista as pv
import numpy as np

# Create a plotter object
plotter = pv.Plotter()

# Define the space boundaries
x_min, x_max = -5, 5
y_min, y_max = -5, 5
z_min, z_max = -5, 5

# Function to plot a sphere (representing asteroids)
def plot_sphere(center, radius, color):
    sphere = pv.Sphere(radius=radius, center=center)
    plotter.add_mesh(sphere, color=color)


num_asteroids = 8 
# # Plot the asteroids
# asteroid_centers = np.array([
#     [1.9, 0, 1.9],
#     [0, 0, -1],
#     [-1.6, 0, 0.5],
#     [0, 0.5, 1],
#     [1, 0, 0.5],
#     [-1.4, 0, -1.4],
#     [0,0,0],
#     [0,0,-3]
# ])
# Create asteroid centers

asteroid_centers = np.vstack([
    np.random.uniform(x_min, x_max, num_asteroids),
    np.random.uniform(y_min, y_max, num_asteroids),
    np.random.uniform(z_min, z_max, num_asteroids),
]).T

asteroid_radii = [0.5, 0.6, 0.4, 0.7, 0.3, 0.4, 0.6, 0.9]
for center, radius in zip(asteroid_centers, asteroid_radii):
    plot_sphere(center, radius, 'brown')

# Function to add an image on a plane
def add_image_on_plane(image_path, center, scale, direction = (0,0,1)):
    img_texture = pv.read_texture(image_path)
    plane = pv.Plane(center=center, direction=direction, i_size=scale, j_size=scale)
    plotter.add_mesh(plane, texture=img_texture)

# Add the bird and space station images
add_image_on_plane('rocket-ship.png', center=[5,5,0], scale=1, direction = (1,1,1))
add_image_on_plane('space-station.png', center=[-4, -4, 4], scale=2)

# Plot stars
stars_x = np.random.uniform(-6, 6, size=200)
stars_y = np.random.uniform(-6, 6, size=200)
stars_z = np.random.uniform(-6, 6, size=200)
stars = pv.PolyData(np.c_[stars_x, stars_y, stars_z])
plotter.add_mesh(stars, render_points_as_spheres=True, color='white', point_size=5)

# Environment Setup
plotter.set_background('#000015')  # Dark background to simulate space

# Display the plot
plotter.show()


# Export the visualization to an image file
filename = "3d_rocket.png"
plotter.screenshot(filename)

plotter.close()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Function to plot a sphere
def plot_sphere(ax, center, radius, color):
    u = np.linspace(0, 2 * np.pi, 100)
    v = np.linspace(0, np.pi, 100)
    x = center[0] + radius * np.outer(np.cos(u), np.sin(v))
    y = center[1] + radius * np.outer(np.sin(u), np.sin(v))
    z = center[2] + radius * np.outer(np.ones(np.size(u)), np.cos(v))
    ax.plot_surface(x, y, z, color=color, alpha=0.5)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# # Plot the main sphere
# plot_sphere(ax, [0, 0, 0], 1, 'b')

# Plot the rocket ship
# Plot initial position of the bird
BIRD_SIZE = 0.05

xmin, xmax = -1-BIRD_SIZE, -1+BIRD_SIZE
ymin, ymax = -1-BIRD_SIZE, -1+BIRD_SIZE
zmin, zmax = -1-BIRD_SIZE, -1+BIRD_SIZE
rocket = ax.imshow(img, extent=[xmin,xmax,ymin,ymax, zmin, zmax], aspect='auto', zorder=2)
# rocket_center = [-1, -1, -1]
# rocket_radius = 0.1
# plot_sphere(ax, rocket_center, rocket_radius, 'r')

# Plot the space station
station_center = [1, 1, 1]
station_radius = 0.1
plot_sphere(ax, station_center, station_radius, 'g')

# Plot the asteroids
num_asteroids = 4
asteroid_colors = ['m', 'y', 'c', 'r']
asteroid_radii = [0.4, 0.2, 0.3, 0.6]
asteroid_centers = [
    [-0.5, 1, 0.5],
    [0, -0.5, 0.5],
    [0, 0, -0.5],
    [1,0,0]
]
for i in range(num_asteroids):
    plot_sphere(ax, asteroid_centers[i], asteroid_radii[i], asteroid_colors[i])

# Plot the curve tangent to the sphere
t = np.linspace(0, 1, 1000)
x_curve = np.sin(np.pi * t)
y_curve = np.cos(np.pi * t)
z_curve = np.cos(np.pi * t)
ax.plot(x_curve, y_curve, z_curve, color='r')

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Scene with Spheres')

plt.show()
```
