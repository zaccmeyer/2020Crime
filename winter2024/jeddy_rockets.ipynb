```python
def obstacle(x,y,z, W1=1,r=(.2,.2,.2),c=(.5,.5,.5)):
    '''
    Define an area that will represent an obstacle
    
    Parameters:
        x (float): x position in space
        y (float): y position in space
        W1 (float): weight of cost
        r (tuple): radius in x and y direction
        c (tuple): center of the ellipse
    '''
    # print(r)
    #distance_squared = ((x - c[0])**2 + (y - c[1])**2 + (z - c[2])**2)
    ellipse = ((x - c[0])**2/r[0] + (y - c[1])/r[1] + (z - c[2])**2/r[2])**20 + 1

    return W1 / ellipse

def obstacle_dx(x,y,z,W1=1,r=(1/5,1/5,1/5),c=(1/2,1/2,1/2)):
    '''
    x derivative of the obstacle

    Parameters:
        x (float): x position in space
        y (float): y position in space
        W (float): weight of cost
        r (tuple): radius in x and y direction
        c (tuple): center of the ellipse
    '''
    # print(r)
    sphere = (x - c[0])**2/r[0]**2 + (y - c[1])**2/r[1]**2 + (z - c[2])**2/r[2]**2
    numer = -40*W1*(sphere)**19 * (x-c[0])
    denom = r[0]**2*((sphere)**20 + 1)**2
    return numer / denom

def obstacle_dy(x,y,z,W1=1,r=(.2,.2,.2),c=(.5,.5,.5)):
    '''
    y derivative of the obstacle

    Parameters:
        x (float): x position in space
        y (float): y position in space
        W1 (float): weight of cost
        r (tuple): radius in x and y direction
        c (tuple): center of the ellipse
    '''
    # print(r)
    sphere = (x - c[0])**2/r[0]**2 + (y - c[1])**2/r[1]**2 + (z - c[2])**2/r[2]**2
    numer = -40*W1*(sphere)**19 * (y-c[1])
    denom = r[1]**2*((sphere)**20 + 1)**2
    return numer / denom

def obstacle_dz(x,y,z,W1 = 1, r= (1/5,1/5,1/5), c = (1/2,1/2,1/2)):
    sphere = (x - c[0])**2/r[0]**2 + (y - c[1])**2/r[1]**2 + (z - c[2])**2/r[2]**2
    numer = -40*W1*(sphere)**19 * (z-c[2])
    denom = r[2]**2*((sphere)**20 + 1)**2
    return numer/denom
```

```python
def ode(t, y_val): #define the ode systen
    x, y,z,xp,yp,zp,p0,p1,p2,p3,p4,p5 = y_val
    # print(np.shape(x))
    # print(np.shape(p))
    new_p0 = obstacle_dx(x,y,z,W)
    new_p1 =  obstacle_dy(x,y,z,W)
    new_p2 = obstacle_dx(x,y,z,W)
    return np.array([xp, yp, zp, p3, p4,p5, new_p0,new_p1,new_p2, -p0, -p1,-p2])
```

```python
def bc(ya,yb):
# ya are the initial values
# yb are the final values
    return np.array([ya[0],ya[1],ya[2],ya[3],ya[4]-3,ya[5]-3,yb[0]-1,yb[1]-1,yb[2]-1,yb[3],yb[4],yb[5]])
```
